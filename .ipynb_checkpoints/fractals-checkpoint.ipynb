```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

# Barnsley Fern Diagram

```python
from ferm import draw_ferm
n = int(input('Enter the number of points in the Barnsley Fern Diagram: '))
x, y = draw_ferm(n)
plt.plot(x, y, 'o', markersize=0.5, color='green')
plt.title('Barnsley Fern Diagram with {} points'.format(n))
plt.show()
```

# Julia Set

```python
from julia import initialyzeplane, julia
n = 1000
print('Draw Julia Set iterating f(z) = z^2 + c')
print('Good values for c are:')
print('1) c = -0.8 + 0.156i')
print('2) c = 0.285 + 0.01i')
print('3) c = -0.4 + 0.6i')
print('4) c = -0.390541 -0.586788i (Siegel Disk Fractal)')
print('5) c = -0.123 + 0.745i (Douady\'s Rabbit Fractal)')
print('6) c = -0.75 (San Marco Fractal)')
print('7) c = -0.8i')
print('8) c = i (Dendrite Fractal)')
print('\n')
real = float(input('Enter real part of c: '))
imag = float(input('Enter imaginary part of c: '))
c = complex(real, imag)
plane = initialyzeplane(n)
fig = julia(plane, c)
plt.imshow(fig, extent=(-2.,2.,-2.,2.), cmap=cm.nipy_spectral_r, interpolation='nearest')
plt.title('Julia Set f(z) = z^2 + c with c = {} + {}i'.format(c.real, c.imag))
plt.colorbar()
plt.show()
```

# Mandelbrot Set

```python
from mandelbrot import initialyzeplane_man, mandelbrot
n = 400
plane = initialyzeplane(n)
fig = mandelbrot(plane)
plt.imshow(fig, extent=(-2.5,1.0,-1.0,1.0), cmap=cm.Greys_r, interpolation='nearest')
plt.title('Mandelbrot Set')
plt.colorbar()
plt.show()
```

# Generalized Mandelbrot Set

```python
from genarlizedMandelbrot import initialyzeplane_mangen, mandelbrot2
n = 400
plane = initialyzeplane_mangen(n)
order = int(input('Insert order of Generalized Mandelbrot Set: '))
fig = mandelbrot2(plane, order)
plt.imshow(fig, extent=(-2.5,1.0,-1.0,1.0),
cmap=cm.Greys_r, interpolation='nearest')
plt.title('Generalized Mandelbrot Set with order {}'.format(order))
plt.colorbar()
plt.show()
```

# Sierpinsky Triangle

```python
from sierpinsky import draw_sierpinsky
n = int(input('Enter the number of points in the Sierpinsky Triangle: '))
x, y = draw_sierpinsky(n)
# Plot points
plt.plot(x, y, 'o', markersize=0.5)
plt.title('Sierpinsky Triangle with {} points'.format(n))
plt.show()
```

# Chaos Game generated fractals

```python
from chaos import chooseNextVertex1, chooseNextVertex2, chooseNextVertex3, chaos
iteration = 100000
# Vertices of a square
vertices = {'A':(0.,0.), 'B':(2.,0.), 'C':(2., 2.), 'D':(0., 2.)}
# Choose a next-vertex rule
function = [chooseNextVertex1, chooseNextVertex2, chooseNextVertex3]
print('[1] Next vertex must be different from previously chosen vertex')
print('[2] Next vertex cannot be 2 places away from the previously chosen vertex')
print('[3] Next vertex cannot be 2 places away from the two previously chosen vertices.')
choice = int(input('Enter next vertex rule: '))
x, y = chaos(vertices, iteration, function[choice-1])
plt.plot(x, y, 'o', markersize=0.5, color='blue')
plt.show()
```
